```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  
import folium as fm
%matplotlib inline 
sns.set(color_codes=True)
```

```python
file_path = "C:\\Users\\praty\\OneDrive\\Desktop\\election-result.xlsx"
df = pd.read_excel(file_path, header=0)
```

```python
df.head()
```

```python
 df['Winning_party'].unique()
```

```python
name_counts = df['Winning_party'].value_counts() 
plt.figure(figsize=(10, 5))
plt.gca().set_facecolor('#F2F2F2')    
colors = ['#008000', '#FFA500', '#000000', '#708090', '#FF0000', '#800080', '#377EB8', '#6495ED', '#FF1493']
plt.bar(name_counts.index, name_counts.values, color=colors) 
plt.grid(True, axis='y', linestyle='--', color='white') 
plt.rcParams['font.family'] = 'Arial' 
plt.ylabel('Wins',fontweight='bold')
plt.title('Party Wins: First-Past-the-Post Election Results',fontweight='bold') 
plt.xticks(rotation=90) 
plt.tight_layout()
plt.show()
```
